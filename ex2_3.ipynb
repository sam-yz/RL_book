```python
import math
import numpy as np

action_set = [1,2]
q = np.random.normal(0, 1, 2)
q[0] = q[0] if q[0] > 0 else -q[0]
q[1] = q[1] if q[1] > 0 else -q[1]
print("q", q)

Qt = {1:[], 2:[]}

optim_reward = 1 if q[0] > q[1] else 2

cum_reward = []
cum_avg = []

tot_iter = 2500
tao = 1

def Q_fn(a):
    if len(Qt[a]) < 1:
        return 0
    return sum(Qt[a])/len(Qt[a])

def proba(a):
    numer = math.exp(Q_fn(a)/tao)
    denom = math.exp(Q_fn(1)/tao) + math.exp(Q_fn(2)/tao)
    return numer/denom

def choose_action():
    U = np.random.uniform(0, 1)
    P = proba(1)
    print("U P |", U, P)
    if U > P:
        return 2
    else:
        return 1

def reward_fn(a):
    """a is the action taken"""
    return (q[0] + np.random.normal(0, 1)) if a == 1 else (q[1] + np.random.normal(0, 1))
    
probs = []
best1 = []
# Run the program

for i in range(0, tot_iter):
    At = choose_action()
    print("Action: ", At)
    Rt = reward_fn(At)
    Qt[At].append(Rt)
    cum_reward.append(Rt)
    p1 = p2 = 0
    if len(Qt[1]) > 0:
        p1 = sum(Qt[1])/len(Qt[1]) #cum avg p1 reward
    if len(Qt[2]) > 0:
        p2 = sum(Qt[2])/len(Qt[2]) #cum avg p2 reward
    cum_avg.append(p1*proba(1) + p2*proba(2)) #expected reward
    probs.append((proba(1), proba(2)))
    best1.append(q[optim_reward-1])
```

```python
import matplotlib.pyplot as plt

plt.plot(cum_avg)
plt.plot(best1)
plt.show()
```
